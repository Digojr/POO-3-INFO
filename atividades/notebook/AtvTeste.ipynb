Gere uma saída de texto entre aspas duplas, criando uma nova linha e posicionando o cursor na linha abaixo.

```java
//Closse impressão de texto:PrimeiroProgramaJava.Java
/*IFPR - Campus Cascavel
Disciplina: Programação Orientação à Obijetos
Professor: Nelson Bellincanta*/

//inicio da class TextoEntreAspass
public class TextoEntreAspas {
    private String texto;// declaração do atributo privado texto

    public TextoEntreAspas(String texto) {// declaração do construtor que recebe uma String como parâmetro
        this.texto = texto; // atribuição do valor do parâmetro ao atributo texto da classe
    }

    public void imprimir() {// declaração do método público imprimir, que não recebe nenhum parâmetro
        System.out.println("\n"+'"'+ texto +'"' + "\n");// gera a saída do texto formatado, usando aspas duplas e criando uma nova linha
    
    }

}
```

```java
//chama a class TextoEntreAspas e a executa    
TextoEntreAspas texto = new TextoEntreAspas("Exemplo de texto entre aspas.");
    texto.imprimir();
```
